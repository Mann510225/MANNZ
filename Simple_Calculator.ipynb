```python
# create a simple calculator using functions for addition, subtraction, multiplication and division of two numbers.
### STEPS

# A) Functions for Operations / Def & return to freeze the action

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

 #B) User interface pf print commands to execute each starting from select

print("Select operation.")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

# C)  Input command to enter choice with 2 nums input and represent as floats

choice = input("Enter choice: ")

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# D) Put conditional logic with options for each execution command

if choice == '1':
    print(f"{num1} + {num2} = {add(num1, num2)}")

elif choice == '2':
    print(f"{num1} - {num2} = {subtract(num1, num2)}")

elif choice == '3':
    print(f"{num1} * {num2} = {multiply(num1, num2)}")

elif choice == '4':
    print(f"{num1} / {num2} = {divide(num1, num2)}")

else:
    print("Invalid input")
```
